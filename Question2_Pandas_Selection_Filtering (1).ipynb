Task 1: Create Employee DataFrame
Create a pandas DataFrame using the employee data.

```python
import pandas as pd
```

```python
import pandas as pd

data = {
    "EmployeeID": [201,202,203,204,205,206,207,208,209,210],
    "Name": ["Sarah","Michael","Jennifer","David","Lisa","James","Emma","Robert","Sophia","William"],
    "Department": ["Engineering","Sales","Marketing","Engineering","HR","Sales","Engineering","Marketing","Sales","HR"],
    "Salary": [85000,62000,71000,92000,58000,68000,78000,75000,65000,60000],
    "Performance Score": [4,3,5,5,4,4,3,4,5,3],
    "Years of Service": [5,3,7,8,4,6,2,5,4,3]
}
df = pd.DataFrame(data)
# Set EmployeeID as index
df.set_index("EmployeeID", inplace=True)
df
```

Task 2: Select Specific Columns
Select and display only the 'Name', 'Department', and 'Salary' columns

```python
df[["Name", "Department", "Salary"]]
```

Task 5: Filter and display all employees from the 'Engineering' department

```python
engineering_employees = df[df["Department"] == "Engineering"]
engineering_employees
```

Task 6: Filter employees with Salary greater than 70000

```python
high_salary = df[df["Salary"] > 70000]
high_salary
```

Task 7: Filter employees from 'Sales' department with Performance Score ≥ 4

```python
sales_high_perf = df[(df["Department"] == "Sales") & (df["Performance Score"] >= 4)]
sales_high_perf
```

Task 8: Sort DataFrame by Salary (Descending) and Show Top 3 Highest-Paid Employees

```python
top_3_highest = df.sort_values(by="Salary", ascending=False).head(3)
top_3_highest
```

Task 9: Sort by Department (Ascending) and Salary (Descending)

```python
sorted_df = df.sort_values(by=["Department", "Salary"], ascending=[True, False])
sorted_df
```
